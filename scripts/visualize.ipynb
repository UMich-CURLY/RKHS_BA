```python
import plotly.graph_objs as go
import pandas as pd

# Read the CSV file
df = pd.read_csv('/home/link/Projects/curly_slam/data/ethz/local_frame/Hokuyo_34.csv')

# filter out the points that 2DscanId!=0 or 1
df = df[df['2DscanId'] < 10]

# Create a scatter plot
fig = go.Figure(data=go.Scatter3d(
    x=df['x'],
    y=df['y'],
    z=df['z'],
    mode='markers',
    marker=dict(
        size=2,
        color=df['Intensities'],  # Use intensity as color
        colorscale='Viridis',  # Set the colorscale
        opacity=0.8
    )
))

# Set the axis labels
fig.update_layout(scene=dict(
    xaxis_title='X',
    yaxis_title='Y',
    zaxis_title='Z'
))

# Show the plot
fig.show()
```
